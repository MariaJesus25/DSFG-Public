```python
def tau(z,b,M_,d_,vf,v0): 
    M=Msun * 10** M_
    k0=0.15*(un.m**2/un.kg) 
    tau=k0*(vf/v0)**b *(z+1)**b * M/(np.pi*(d_/2)**2)
    return tau.to("")
```

```python
def P(vf,T,z):
    h=cte.h
    K=cte.k_B
    A=2*h/(c**2)
    a=(h*vf*(1+z)/(K*T))
    B=vf**3/(np.e**(a.value) -1)
    P=(A*B).to("mJy")
    return P
```

```python
def model(z,T,b,M_,d_,vf,v0,u):
    D=cosmo.angular_diameter_distance(z)
    print("Angular diameter distance:{}".format(D.to("Mpc")))
    M=Msun*(10**M_)
    d=d_/(u**1/2)
    O=np.pi*(d/2)**2 * (D**-2)
    print(tau(z,b,M_,d_,vf,v0))
    S_=O*(1-np.e**-tau(z,b,M_,d_,vf,v0))*P(vf,T,z)
    S=S_.to("mJy")
    return S
```
