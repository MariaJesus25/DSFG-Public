```python
def tau(z,b,M_,u,d_,v_,v0): #asumiendo que la Galaxia se comporta como un BB
    #k0 mass opacity coefficient referred to v0
    v=c_/v_
    M=(Msun*(10**M_))/u
    d=d_/(u**1/2)
    k0=0.15*un.m**2/un.kg
    tau= k0 * (v/v0)**b * (1+z)**b * M/(np.pi*(d/2)**2)
    return tau
```

```python
print(tau(3.5,2,9.45,1,2.9*un.kpc, 100*un.um,353*un.GHz).to(""))
```

```python
def P(v,T,z):
    h=cte.h
    K=cte.k_B
    A=2*h/(c_**2)
    a=(h*v*(1+z)/(K*T))
    B=v**3/(np.e**(a.value) -1)
    P=(A*B).to("mJy")
    return P
```

```python
print(P(vf*un.Hz,45.9*un.K,3.5))
```

```python
def model(z,T,b,M_,d_,u,P,tau):
    LM=0.27
    LL=0.73
    H0=70*un.km/(un.s*un.Mpc)
    D=cosmo.angular_diameter_distance(z)
    print("Angular diameter distance:{}".format(D.to("Mpc")))
    M=(Msun*(10**M_))/u
    d=d_/(u**1/2)
    O=np.pi*(d/2)**2 * (D**-2)
    print(O.to(""))
    S_=O*(1-np.e**-tau)*P
    S=S_.to("mJy")
    return S
```

```python
model(3.5,45.9*un.K,2,9.45,2.9*un.kpc,1.6,P(vf*un.Hz,45.9*un.K,3.5),1)
```
